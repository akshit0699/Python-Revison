```python
# 1. GIVEN TWO TREES, ARE THEY STRUCTURALLY IDENTICAL AND HAVE SAME NODE VALUES?
# rec case, if the value of the two heads is same, and the left subtree and the right subtree are both same
# base case, if both of them are NULL toghter, yes, If one becomes null othe does not, no!

def isSame(p,q):
    if not p and not q:
        return True
    if not p or not q:
        return True
    return p.val == q.val and isSameTree(p.left, q.left) and isSameTree(p.right, q.right) 
    
```

```python
# 2. GIVEN A TREE, IS IT SYMMETRIC AROUND ITS CENTER? IS THE RIGHT PART A MIRROR IMAGE OF THE LEFT PART?
# rec case, if the value of the two heads is same, and the left subtree and the right subtree are both symmetric
# Cant we simple say if left and right subtrees are independent trees,
# if left of leftsubtree is same as right of righsubtree and the other way around is also true, then Done!
def isSymmetric(root):
    if not root:
        return False
    helper(root.left, root.right)
    def helper(p, q):
        if not p and not q:
            return True
        if not p or not q:
            return False
        return p.val == q.val and helper(p.left, q.right) and helper(p.right, q.left)
        
```

```python
# GIVEN A TREE, FIND ITS MAX DEPTH
def maxDepth(root):
        if not root:
            return 0
        lh = self.maxDepth(root.left)
        rh = self.maxDepth(root.right)
        return 1 + max(lh, rh)
        
```

```python
# GIVEN A BINARY TREE, PRINT ALL PATHS FROM ROOT TO LEAF
def allPaths(root):
    if not root:
        return []
    res = []
    helper(root, "", res)
    return res
def helper(root, path, res):
    if root.left is None and root.right is None:
        res.append(path+str(root.val))
    if root.left:
        helper(root.left, res, path+str(root.val)+"->")
    if root.right:
        helper(root.right, res, path+str(root.val)+"->")
        
```

```python
# IS THERE A ROOT TO LEAF PATH THAT HAS SUM = TARGET?
def haspathSum(root, target):
    if not root:
        return False
    if root.left is None and root.right is None and target == root.val:
            return True
    canLeft = self.pathSum(root.left, target-root.val)
    canRight = self.pathSum(root.right, target- root.val)
    return canLeft or canRight    
```

```python
# Simple, Invert the god damn tree, make the lefts right
def invertTree(root):
    if not root:
        return None
    temp = invertTree(root.left)
    root.left = invertTree(root.right)
    root.left = temp
    return root
    
```

```python
def invertTreeOPT(root):
    if not root:
        return None
    queue = collections.deque([root])
    while queue:
        node = queue.popleft()
        if node:
            # Same as using a temp, (More pythonic way )
            node.left, node.right = node.right, node.left
            queue.append(node.left)
            queue.append(node.right)
        return root
```

```python
# A node can be considered to be the LCA of two nodes, if it has p in left subtree, and q in right subtree
# Then that is node
# If I do not find p/q in left, but nothing in right. Search only in left
# vice versa for right

def lowestCommonAncestor(root, p,q):
    if not root:
        return None
    if p == root or q == root:
        return root
    left = self.lowestCommonAncestor(root.left, p,q)
    right = self.lowestCommonAncestor(root.right,p,q)
    
    # I've got both the left and right children! This is the result!
    if left and right:
        return root
    
    if not left:
        return right
    if not right:
        return left
  # either one of the chidren returned a node, meaning either p or q found on left or right branch.
  # Example: assuming 'p' found in left child, right child returned 'None'. This means 'q' is
  # somewhere below node where 'p' was found we dont need to search all the way, 
  # because in such scenarios, node where 'p' found is LCA  
    
```

```python
# If my left subtree is balanced, right is also balanced, and I have checked my condititon, then its all GOOOOOOD!
def isBalanced(root):
    if not root:
        return True
    lh = height(root.left)
    rh = height(root.right)
    if abs(lh-rh)<=1 and self.isBalanced(root.left) and self.isBalanced(root.right):
        return True
    return False

def height(root):
    if root is None:
        return 0
    return 1 + max(self.height(root.left), self.height(root.right))
    
    
```

```python
# SEREALIZE AND DESERIALIZE A TREE

################SERAIALIZATION########################

     # The key to tree problems is TO HAND OFF RESPONSIBILITY. Here we are saying...
     # 
     # "hey, I know what root's value is...ummmm...hey serialize()... can you encode
     # my left and right subtrees?...Get back to me because ONLY THEN will I append
     # myself"
     #
# I am given a tree, I need to convert it into an array
# Base case: if its a null, return 'x'
# Else append, root.val with the left sides answer and the right sides answer, all joined by a comma.
# ','.join([any list])
        # preoder(Root->left->right)
        # ex,
        #     1
        #    / \
        #   2   3
        #      / \
        #     4   5 
        # 
        # return (1, (2, 'x', 'x'), (3, (4, 'x', 'x'), (5, 'x', 'x')))
        # if you look at the return statement very closely, it is actually very intuitive
        # for value 1, you have 2 as left child and 3 as right child
        # for value 2, you have 'x'(None) as left child and 'x'(None) as right child which indicates it is a leaf node
        
################DESERAIALIZATION########################
# Now that array has to be converted back into a tree.
# if you see a 'x' return None
# Other wise create a node of value till the index of ,
# find this node's left guy, send the remainder list
# find this node's right guy, send the remainder list
# Return the node.
# pull of the first item, make a node out of it, and then we go left and and left and left untill we hit a x,
        # The initial data string will be something like below:
        # (1, (2, 'x', 'x'), (3, (4, 'x', 'x'), (5, 'x', 'x')))
        # if you loop through string: 
        # 1                                 -> this is node value
        # (2, 'x', 'x')                     -> this is node left
        # (3, (4, 'x', 'x'), (5, 'x', 'x')) -> this is node right
```

```python
class Codec:
    def serialize(self, root):
        if not root: return 'x'
        return ','.join([str(root.val), self.serialize(root.left), self.serialize(root.right)])
        
    def deserialize(self, data):
        self.data = data
        if data[0] == 'x': return None
        node = TreeNode(self.data[:self.data.find(',')]) 
        node.left = self.deserialize(self.data[self.data.find(',')+1:])
        node.right = self.deserialize(self.data[self.data.find(',')+1:])
        return node
```
