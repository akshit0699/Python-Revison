```python
# Given a range m to n, you have to reverse in that range only
def reverseRange(head, m, n):
        
        dummy = ListNode(0)
        save = dummy
        dummy.next = head
        for _ in xrange(m-1):
            save = save.next
        # So we have our savepoint just behind the reversing point
        pre = None
        if save.next:
            curr = save.next
        # Reverse as per requirements
        for _ in xrange(n-m+1):
            temp = curr
            curr = curr.next
            temp.next = pre
            pre = temp
        
        # Join them all togehter.
        save.next.next = curr
        save.next = pre
        return dummy.next
```

```python
def removeDuplicates(head):
    while curr and curr.next:
        if not head or not head.next:
            return head
        curr = head
        if curr.val == curr.next.val:
            curr = curr.next
        else:
            curr.next = curr.next.next
            
def removeDuplicatesFaster(head):
        if not head or not head.next:
            return head
        temp = head
        while temp.next and temp.next.next:
            if temp.val == temp.next.val:
                temp.next = temp.next.next
            else:
                temp = temp.next
        if temp.val == temp.next.val:
            temp.next = None
        return head
        
```

```python
'''
We have to tell if the values represnted by a ll form a palindrome or not
1523251 --> len = 7
Go to the middle node and reverse the stuff after it
After that start traversing both, if the both come out to be the same
its a palindrome
Take care if the lenght is odd, then skip ahead one while keeping the pointer for right side
'''
def palidrome(head):
    fast, slow = head, head
    while fast and fast.next:
        fast = fast.next.next
        slow = slow.next
    
    # We will now be reversing with slow == curr
    node = None
    while slow:
        temp = slow
        slow = slow.next
        temp.next = node
        node = temp
    while node:
        if node.val != head.val:
            return False
        node = node.next
        head = head.next
    return True

def isPalindrome(self, head):

    if not head or not head.next:
        return True

    # 1. Get the midpoint (slow)
    slow = fast = cur = head
    while fast and fast.next:
        fast, slow = fast.next.next, slow.next
    
    # 2. Push the second half into the stack
    stack = [slow.val]
    while slow.next:
        slow = slow.next
        stack.append(slow.val)

    # 3. Comparison
    while stack:
        if stack.pop() != cur.val:
            return False
        cur = cur.next
    
    return True
```

```python
'''
Simple obvious method: Find the length of the list remove (len - n + 1)element from the list 
Trying to do it in a single pass! Not super efficient though, but worth knowing

'''
def removeNthFromEnd(head, n):
    temp = head
    count = 0
    while temp:
        count+=1
        temp = temp.next
    n = n - count # Taking up from the front now
    if n == 0:
        return head.next
    temp = head
    while n>1 and temp:
        n-=1
        temp = temp.next
    temp.next = temp.next.next
    return head

def removeNthNodeSinglePasss(head, n):
    dummy = ListNode(0)
    first = second = dummy
    dummy.next = head
    
    # Make the first pointer n steps ahead
    for _ in xrange(n):
        first = first.next
    
    # Move both till the first reaches a dead end
    while first and first.next:
        first = first.next
        second = second.next
    second.next = second.next.next
    return dummy.next
```

```python
'''
Maintain two pointers pApA and pBpB initialized at the head of A and B, respectively.
Then let them both traverse through the lists, one node at a time.
When pApA reaches the end of a list, then redirect it to the head of B (yes, B, that's right.)
Similarly when pBpB reaches the end of a list, redirect it the head of A.
'''

def intersectionPoint(headA, headB):
        if headA == None or headB == None:
            return None
        tempA, tempB = headA, headB
        while tempA!=tempB:
            if not tempA or not tempB:
                if not tempA:
                    tempA = headB
                if not tempB:
                    tempB = headA
            else:
                tempA = tempA.next
                tempB = tempB.next
        return tempA
    
```

```python
class Solution(object):
    def hasCycle(self, head):
        """
        :type head: ListNode
        :rtype: bool
        Is there is a cycle in the linked list..
        Keep a slow pointer and a fast pointer
        If they converge at any point there is a cycle!
        """
        if not head or not head.next:
            return False
        
        slow = head
        fast = head
        while slow and fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                return True
        return False
```

```python
class Solution(object):
    def sortList(self, head):
        """
        :type head: ListNode
        :rtype: ListNode
        We keep breaking our list into two parts and then we keep merging them
        Typical merge sort divide and conquer is followed
        """
        if not head or not head.next:
            return head
        slow = head
        fast = head.next
        while fast and fast.next:
            fast = fast.next.next
            slow = slow.next
        fast = slow.next
        slow.next = None
        lsorted = self.sortList(head)
        rsorted= self.sortList(fast)
        return self.merge(lsorted, rsorted)
        
    def merge(self, l1, l2):
        if not l1 or not l2:
            return l1 or l2
        
        head = None
        if l1.val<=l2.val:
            head = l1
            l1 = l1.next
        else:
            head = l2
            l2 = l2.next
        dummy = head
            
        while l1 and l2:
            if l1.val<=l2.val:
                head.next = l1
                l1 = l1.next
            else:
                head.next = l2
                l2 = l2.next
            head = head.next
                
        if l1:
            head.next = l1
        if l2:
            head.next = l2
        
        return dummy
        
        
```
