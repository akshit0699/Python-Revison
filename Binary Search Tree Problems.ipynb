```python
def lowestCommonAncestor(self, root, p, q):
    while root:
        if root.val > p.val and root.val > q.val:
            root = root.left
        elif root.val < p.val and root.val < q.val:
            root = root.right
        else:
            return root
```

```python
class Solution(object):
    def recoverTree(self, root):
        """
        :type root: TreeNode
        :rtype: None Do not return anything, modify root in-place instead.
        """
        self.first, self.second, self.prevNode = None, None, None # Create three nodes.
        self.inOrder(root) # Calling the function
        self.first.val, self.second.val = self.second.val, self.first.val 
        # Swapping the two elements needed to be swapped
        
    def inOrder(self, root):
        if not root:
            return
        self.inOrder(root.left)
        # DO BUSINESS!!!!!
        if self.prevNode: # To handle the case of first node, because we make it prev to begin with
            if self.prevNode.val > root.val: # Check property violation
                if not self.first: 
                    self.first = self.prevNode # Found first pair
                self.second = root # If the second pair is found then simply assign the smaller element of the   pair as the second guy, it works for single pair easily, as it wont get             updated again in that case.
                
        self.prevNode = root
        
        self.inOrder(root.right)
        
```

```python
# To search the node's position, first step is to know whether it is in the left subtree or right subtree
# We will MAKE THAT NODE ROOT!! and then delete!
# Easy case: node has no children, delete it simply
# Also Easy: node has only one child.
# Tough one: NODE HAS BOTH CHILDREN!!

def deleteNode(root, key):
    if root is None:
        return None
    if root.val > key:
        root.left = self.deleteNode(root.left, key)
    elif root.val < key:
        root.right = self.deleteNode(root.right, key)
    else:
        if not root.right: # Also easy (1)
            return root.left # Easy case covered
        if not root.left: # Also easy (2)
            return root.right
        
        # There is a left and right child both,go right and then keep going left(inorder successor) can also go for predecessor
        inorderSucc = root.right
        while inorderSucc.left:
            inorderSucc = inorderSucc.left
        inorderSucc.left = root.left # make its left as the root's left!
        #return right child instead of root a.k.a deleted root
        return root.right # return root's right, we got rid of root.
    return root
        
```
